```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from random import randint
from scipy.integrate import quad
```

```python
ilość=randint(1,30)
l=[]
przedziały=5/(ilość*2)
for i in range(ilość*2+1):
    l.append(przedziały*i)
```

```python
lewy_róg=[]
for i in range(0,ilość*2,2):
    lewy_róg.append(2/(3+2*l[i]**2))

prawy_róg=[]
for i in range(2,ilość*2+1,2):
    prawy_róg.append(2/(3+2*l[i]**2))
    
x_kolumn=[]
for i in range(1,ilość*2,2):
    x_kolumn.append(l[i])
    
szerokość=5/ilość        
```

```python
x_prawy_róg=[]
for i in range(2,ilość*2+1,2):
    x_prawy_róg.append(l[i])
    
x_lewy_róg=[]
for i in range(0,ilość*2,2):
    x_lewy_róg.append(l[i])
    
suma1=0
for i in range(ilość):
    suma1=suma1+szerokość*prawy_róg[i]
suma1=round(suma1,6)   

suma2=0
for i in range(ilość):
    suma2=suma2+szerokość*lewy_róg[i]
suma2=round(suma2,6)  
suma2


from scipy.integrate import quad
def f(x):
    return 2/(3+2*x**2)


a, b = quad(f, 0, 5)
pole_całki=a-b
pole_całki=round(pole_całki,6)
```

```python
x = np.linspace(0, 5, 400)
y = 2/(3+2*x**2)


fig, (ax1,ax2,ax3) = plt.subplots(1,3,sharex=True,sharey=True,figsize=(10,10))

ax1.plot(x, y,color='black',zorder=2)
ax1.set_title('suma I={}'.format(suma1),size=15)
ax1.bar(x_kolumn, prawy_róg, width=szerokość, edgecolor='white',zorder=1)
ax1.scatter(x_prawy_róg,prawy_róg,color='red',zorder=3)
ax1.set_xlim(0,5)

ax2.fill_between(x,y,color='black')
ax2.plot(x, y,color='red')
ax2.text(1,0.5,'$ f(x)=\dfrac{2}{(3+2x^2)}$',size=15)

ax3.plot(x, y,color='black',zorder=3)
ax3.set_title('suma II={}'.format(suma2),size=15)
ax3.bar(x_kolumn, lewy_róg, width=szerokość, bottom=None, align='center', data=None,edgecolor='white',zorder=1)
ax3.scatter(x_lewy_róg,lewy_róg,color='red',zorder=2)
ax3.set_xlim(0,5)

fig.suptitle("Przedziałów jest:{} " 
             ' \n Pole jest równe:{}'.format(ilość,pole_całki),size=15)

plt.savefig('05_Ort_272194')
plt.show()
```
