```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

```python
x1=random.uniform(0,2)
x2=random.uniform(2,4)
x3=random.uniform(4,6)
x4=random.uniform(6,8)
```

```python
x1,x2,x3,x4
```

```python
fig = plt.figure()
ax=fig.add_subplot()
N = 100000
x = np.linspace(0, 10, N+1)
y=(x-x1)*(x-x2)*(x-x3)*(x-x4)

dydx=np.gradient(y,x)
stacj_x = x[1:][dydx[1:]*dydx[:-1]<0]
stacj_y = y[1:][dydx[1:]*dydx[:-1]<0]
ax.scatter([stacj_x[0], stacj_x[1], stacj_x[2]],[stacj_y[0],stacj_y[1],stacj_y[2]],
           color = 'black')
y1=stacj_y[0]+stacj_y[2]
ax.set_ylim(y1,stacj_y[1]+ 20)

plt.plot(x[x<stacj_x[0]], y[x<stacj_x[0]],color = 'red',label='maleje')
plt.plot(x[x>stacj_x[0]], y[x>stacj_x[0]],color = 'green',label='rośnie')
plt.plot(x[x>stacj_x[1]], y[x>stacj_x[1]],color = 'red')
plt.plot(x[x>stacj_x[2]], y[x>stacj_x[2]],color = 'green')


ax.set_xticks([stacj_x[0], stacj_x[1], stacj_x[2]])
ax.axhline(0, color='red')



v=(stacj_x[1]+stacj_x[0])/2
w=stacj_y[1]-stacj_y[0]
w=round(w,6)
t1=(v-x1)*(v-x2)*(v-x3)*(v-x4)

ax.annotate(w,xy=(v,t1),xytext=(v,stacj_y[0]), 
            size=10,arrowprops=dict(arrowstyle='-'))


v1=(stacj_x[2]+stacj_x[1])/2
w1=stacj_y[1]-stacj_y[2]
w1=round(w1,6)
t2=(v1-x1)*(v1-x2)*(v1-x3)*(v1-x4)


ax.annotate(w1,xy=(v1,t2),xytext=(v1,stacj_y[1]+10), 
            size=10,arrowprops=dict(arrowstyle='-'))

X1=round(x1,6)
X2=round(x2,6)
X3=round(x3,6)
X4=round(x4,6)
ax.set_title('Pierwiastki:{'+str(X1)+'|'+str(X2)+'|'+str(X3)+'|'+str(X4)+'}')
ax.legend(loc='best')

plt.savefig('wykres zad.pdf')
plt.show()
```
