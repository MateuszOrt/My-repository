```python
f= open("Pan-Tadeusz.txt",'r',encoding='utf-8')
text=f.read().split()
f.close()
```

```python
text
```

```python
slowa=list(text)
slowa
```

```python
slowa2=list(map(lambda x: x.lower(),slowa))
slowa2=list(map(lambda x: x.replace(',',''), slowa2))
slowa2=list(map(lambda x: x.replace('.',''), slowa2))
slowa2=list(map(lambda x: x.replace('!',''), slowa2))
slowa2=list(map(lambda x: x.replace('?',''), slowa2))
slowa2=list(map(lambda x: x.replace('...',''), slowa2))
slowa2=list(map(lambda x: x.replace(':',''), slowa2))
```

```python
len(slowa2)
```

```python
slowa3=[]
for i in range(len(slowa2)):
    if len(slowa2[i]) >= 5:
        slowa3.append(slowa2[i])
```

```python
slowa3
```

```python
from collections import Counter
slowa4=Counter(slowa3)
```

```python
slowa4
```

```python
slowa5=list(slowa4.most_common(20))
```

```python
a1=[]
a2=[]
for i in range(20):
    a1.append(slowa5[i][0])
for i in range(20):
    a2.append(slowa5[i][1])
```

```python
a1,a2
```

```python
a1.reverse()
a2.reverse()
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
cmap=plt.get_cmap('summer')


fig=plt.figure(figsize=(18,14))
l=[]
for i in (a2):
    j=(i*0.006)
    l.append(j)



plt.barh(a1,a2,color=cmap(l))

for i in range(len(a1)):
    plt.annotate(a2[i], (a2[i]/2, i-0.2 ),size=15)

for i in range(len(a1)):
    plt.annotate(a1[i], (a2[i]+1, i-0.2 ),size=15)

ax=fig.add_subplot()
ax.set_yticks('')


plt.title('Adam Mickiewicz: Pan Tadeusz{ 69 094 słów}. \n 20 najczęściej występujących słów długości co najmniej 5',size=20)
plt.savefig('04_Ort_272194.svg')
plt.show()
```
