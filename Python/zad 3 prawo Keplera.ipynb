```python
import matplotlib.pyplot as plt
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
```

```python
plt.style.use('dark_background')


def getImage(path):
    return OffsetImage(plt.imread(path))




time = [87.969**2, 224.701**2, 365.256**2, 686.960**2, 4333.287**2, 10756.200**2, 30707.490**2, 60223.353**2]

distance=[57909170**3, 108208926**3, 149597887**3, 227936637**3, 778412027**3, 1426725413**3, 2870972220**3, 4498252900**3]



fig=plt.figure(figsize=(16,14))
ax=fig.add_subplot()

plt.plot([87.969**2, 60223.353**2],[57909170**3,4498252900**3],color='white')

paths = ['Mercury.png','Venus.png','Earth.png','Mars.png','Jupiter.png','Saturn.png','Uranus.png','Neptune.png']


ax.scatter(time,distance)

for time, distance, path in zip(time, distance,paths):
    ab = AnnotationBbox(getImage(path), (time, distance), frameon=False)
    ax.add_artist(ab)
    


ax.set_xscale('log')
ax.set_yscale('log')
ax.set_title('Trzecie prawo Keplera',size =30)
ax.set_xlabel('Okres orbitalny (T)',size=30)
ax.set_ylabel('Pół oś wielka (r)',size=30)
ax.text(0.1,0.6,'$\dfrac{T^3}{r^2}=const$',transform=ax.transAxes,size=20)
plt.savefig('03_Ort_2.pdf')
plt.show()
```
