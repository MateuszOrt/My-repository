```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import urllib.request
```

```python
url = "http://www.mbnet.com.pl/dl.txt"
file=urllib.request.urlopen(url)
wynik=[]

for line in file:
    linia=line.decode("utf-8")
    wynik.append(linia)
```

```python
a=len(wynik)
b=6000
c=a-b
c
```

```python
liczby=[]
```

```python
wynik2=0
for i in range(c,999):
    wynik2=wynik[i][16:len(wynik[i])-2]
    liczby.append(wynik2)   
```

```python
wynik2=0
for i in range(999,a):
    wynik2=wynik[i][17:len(wynik[i])-2]
    liczby.append(wynik2)
```

```python
liczby
```

```python
liczby2=[]
for j in range(6000):
    lista=[]
    i=0
    while len(lista)<6:
   
        if i==len(liczby[j])-1:
            lista.append(liczby[j][i])
            break
        if liczby[j][i] != ',' and liczby[j][i+1] != ',' :
            lista.append(liczby[j][i]+liczby[j][i+1])
            if i+1==len(liczby[j])-1:
                break
            i=i+2
        if liczby[j][i] != ',' :
            lista.append(liczby[j][i])
            i=i+1
        else:
            i=i+1  
         
    liczby2.append(lista)
  
```

```python
from collections import Counter
from itertools import combinations

d  = Counter()
for i in liczby2:
    
    i.sort()
    for j in combinations(i,2):
        d[j] += 1
d.most_common()
```

```python
powtórzenia=[]
wszystkie=0
wszystkie=d.most_common(99999999)

for i in range(len(wszystkie)):
    powtórzenia.append(wszystkie[i][1])
powtórzenia
len(powtórzenia)
a=0
for i in (powtórzenia):
    a=a+i   
    
linia_referencyjna=a/len(powtórzenia)
```

```python
najczęstsze=d.most_common(20)
najczęstsze
```

```python
a1=[]
for i in range(20):
    a1.append(i)

a2=[]
for i in range(20):
    a2.append(int(najczęstsze[i][1]))

a3=[]
for i in range(20):
    a3.append(najczęstsze[i][0])
```

```python
liczby2_int=[]
for i in range(len(liczby2)):
    l=[]
    for j in range(6):
        l.append(int(liczby2[i][j]))
    liczby2_int.append(l) 
        
liczby2_int
```

```python
zero_P=[]
jeden_P=[]
dwa_P=[]
trzy_P=[]
cztery_P=[]
pięć_P=[]
sześć_P=[]
for i in range(len(liczby2_int)):
    p=0
    np=0
    for j in range(6):
        if liczby2_int[i][j]%2==0:
            p=p+1
    if p == 0:
        zero_P.append(liczby2_int[i][j]) 
    if p == 1:
        jeden_P.append(liczby2_int[i][j])
    if p == 2:
        dwa_P.append(liczby2_int[i][j])
    if p == 3:
        trzy_P.append(liczby2_int[i][j])
    if p == 4:
        cztery_P.append(liczby2_int[i][j])
    if p == 5:
        pięć_P.append(liczby2_int[i][j])
    if p == 6:
        sześć_P.append(liczby2_int[i][j])
        
```

```python
l=[]
l.append(len(zero_P)/6000*100)
l.append(len(jeden_P)/6000*100)
l.append(len(dwa_P)/6000*100)
l.append(len(trzy_P)/6000*100)
l.append(len(cztery_P)/6000*100)
l.append(len(pięć_P)/6000*100)
l.append(len(sześć_P)/6000*100)
```
