```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import urllib.request
import seaborn as sns
```

```python
f= open("english3.txt",'r',encoding='utf-8')
text=f.read().split()
f.close()
```

```python
alfabet=['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']
```

```python
pary=[]
for i in alfabet:
    for j in alfabet:
        pary.append(i+j)
```

```python
text2=([i for i in text if len(i) > 1])
```

```python
text3=[]
for i in range(len(text2)):
    text3.append(text2[i][0:2])
```

```python
liczby=[]
for i in pary:
     liczby.append(text3.count(i))
```

```python
dane=[]
for i in range(len(alfabet)):
    dane.append(liczby[i*26:i*26+26])
```

```python
dane2 = pd.DataFrame(dane,index=alfabet,columns=alfabet)
```
